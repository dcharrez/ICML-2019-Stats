```python
import plotly.plotly as py
import plotly.graph_objs as go
import plotly.io as pio

from IPython.display import Image

from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot

# Submission / Acceptance

nAcceptedPapers = [270, 322, 434, 621, 774]
nAcceptanceRates = [26.0, 24.0, 25.9, 25.1, 22.6]
nSubmissions = [1037, 1342, 1676, 2473, 3424]

trace0 = go.Bar(
    x=['ICML 2015', 'ICML 2016', 'ICML 2017', 'ICML 2018','ICML 2019'],
    y= nSubmissions,
    text=nSubmissions,
    textposition = 'auto',
    name='Papers submitted',
    marker=dict(
        color='rgb(49,130,189)'
    )
)
trace1 = go.Bar(
    x=['ICML 2015', 'ICML 2016', 'ICML 2017', 'ICML 2018','ICML 2019'],
    y=nAcceptedPapers,
    text=nAcceptedPapers,
    textposition = 'auto',
    name='Papers accepted',
    marker=dict(
        color='rgb(204,204,204)',
    )
)

trace2 = go.Bar(
    x=['ICML 2015', 'ICML 2016', 'ICML 2017', 'ICML 2018','ICML 2019'],
    y=nAcceptanceRates,
    text=nAcceptanceRates,
    textposition = 'auto',
    name='Papers ',
    marker=dict(
        color='rgb(204,102,0)',
    )
)


data = [trace0, trace1, trace2]
layout = go.Layout(
    xaxis=dict(tickangle=-45),
    barmode='group',
)

init_notebook_mode(connected=True)

fig = go.Figure(data=data, layout=layout)
# plot(fig, filename='angled-text-bar')

static_image_bytes = pio.to_image(fig, format='png')
Image(static_image_bytes)

# pio.write_image(fig, file='acceptance_rate.png', format='png')
```
